```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
# Grouping by species and calculating the mean for each group
grouped_means = df.groupby('species').mean()
grouped_means
```

```python
plt.figure(figsize=(8, 4))
plt.plot(df.index, df['sepal length (cm)'], label='Sepal Length')
plt.title('Sepal Length Over Index')
plt.xlabel('Index')
plt.ylabel('Sepal Length (cm)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x='species', y='petal length (cm)', data=df, estimator='mean')
plt.title('Average Petal Length per Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df['sepal width (cm)'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(df['sepal length (cm)'], df['petal length (cm)'], c=df['species'], cmap='viridis')
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.colorbar(label='Species')
plt.show()
```

```python
# Findings & Observations
# 1. The dataset contains 150 rows and no missing values.
# 2. Each species (0, 1, 2) has distinct average values for petal and sepal features.
# 3. The line chart shows the trend of sepal length across data entries.
# 4. The bar chart shows species 2 has the highest average petal length.
# 5. The histogram indicates most sepal width values are between 2.5 cm and 3.5 cm.
# 6. The scatter plot shows a strong relationship between sepal length and petal length, with species forming clear clusters.
```
